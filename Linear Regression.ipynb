# Simple Linear Regression

Population model: 

$\hspace{2mm} y_i = \beta_0 + \beta_1x_i + \epsilon, \hspace{2mm} \epsilon \overset{iid}{\sim} \mathcal{N}(0, \sigma^2)$

* Generalized form: $\hspace{2mm} y_i | x_i \overset{iid}{\sim} \mathcal{N}(\beta_0 + \beta_1x_i, \sigma^2)$

## Ordinary Least Squares

Minimize Residual Sum-of-Squares:

$\underset{\hat{\beta}_0, \hat{\beta}_1}{\text{argmin}} \space RSS
= \underset{\hat{\beta}_0, \hat{\beta}_1}{\text{argmin}}
\sum\limits^N_{i=1} \left( y_i - (\hat{\beta}_0 + \hat{\beta}_1 x_i) \right)^2$

Normal Equations:

$\frac{\partial RSS}{\partial \hat{\beta}_0} = 0, \space \frac{\partial RSS}{\partial \hat{\beta}_1} = 0$

Solution:
* $\hat{\beta}^{OLS}_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} = \frac{SXY}{SXX}$

* $\hat{\beta}^{OLS}_0 = \bar{y} - \hat{\beta}_1 \bar{x}$

## Maximum Likelihood Estimation

## Estimate Variance

Unbiased estimate for $\sigma^2$:

* $\hat{\sigma}^2 = S^2 = \frac{1}{N-2} \sum\limits_{i=1}^{N} \hat{\epsilon}^2_i, \hspace{2mm}$

## Properties

* $\bar{y} = \hat{\beta}_0 + \hat{\beta}_1 \bar{x}$

* $\sum\limits_{i=1}^N \hat{e}_i = 0, \hspace{6mm} \sum\limits_{i=1}^N \hat{e}_i x_i = 0, \hspace{6mm} \sum\limits_{i=1}^N \hat{e}_i \hat{y}_i = 0, \hspace{6mm} \sum\limits_{i=1}^N \hat{y}_i = \sum\limits_{i=1}^N y_i$

Estimated coefficients are normally distributed with mean 0 (unbiased) and covariance matrix:

* $\begin{bmatrix} \hat{\beta}_0 \\ \hat{\beta}_1 \end{bmatrix} \sim \mathcal{N} \left( \hspace{1mm} \begin{bmatrix} \beta_0 \\ \beta_1 \end{bmatrix}, \hspace{1mm} \begin{bmatrix} \sigma^2 \left( \frac{1}{N} + \frac{\bar{x}^2}{\sum (x_i - \bar{x})^2} \right) & . \\ - \bar{x} \frac{\sigma^2}{\sum (x_i - \bar{x})^2} & \frac{\sigma^2}{\sum (x_i - \bar{x})^2} \end{bmatrix} \hspace{1mm} \right)$

Residuals are correlated:

* $\mathbb{COV}(\hat{\epsilon}_i, \hat{\epsilon}_j) = \begin{cases} -h_{ij}\sigma^2, \hspace{2mm} i \ne j \\ (1 - h_{ii})\sigma^2, \hspace{2mm} i = j \end{cases}, \space$ ($h_{ii}$ is the leverage of i-th data point)

* $h_{ij} = \frac{1}{N} + \frac{(x_i - \bar{x})(x_j - \bar{x})}{SXX}$

* $\sum\limits^N_{j=1} h_{ij} = 1, \hspace{6mm} \sum\limits^N_{j=1} h^2_{ij} = h_{ii}, \hspace{6mm} \hat{y}_i = h_{ii}y_i + \sum\limits_{j \ne i} h_{ij}y_j$

## Inference

* $\frac{(N-2)S^2}{\sigma^2} \sim \chi^2_{(N-2)}$

Tests for $H_0: \beta_1 = 0 \space$ vs $\beta_1 \ne 0$:

* $t_{obs} = \frac{\hat{\beta}_1 - \beta_{1,null}}{SE(\hat{\beta}_1)} = \frac{\hat{\beta}_1 - \beta_{1,null}}{\sqrt{\frac{S^2}{SXX}}} \overset{H_0}{\sim} T_{(N-2)} \implies 100(1 - \alpha)\% \text{ CI}(\beta_1) = \hat{\beta}_1 \pm t^*_{\frac{\alpha}{2},N-2}SE(\hat{\beta}_1)$

* $F_{obs} = \frac{MSReg}{MSE} = \frac{SSReg/1}{SSE/(N-2)} \overset{H_0}{\sim} F_{(1,N-2)}$

## Prediction

$\mathbb{V} \left( \hat{y} \space|\space X=x^* \right) = 
    \sigma^2 \left( \frac{1}{N} + \frac{(x^* - \bar{x})^2}{\sum (x_i - \bar{x})^2} \right) \space$ (for population regression line)
    
$\mathbb{V} \left( \hat{y}^* \space|\space X=x^* \right) = 
    \sigma^2 \left( 1 + \frac{1}{N} + \frac{(x^* - \bar{x})^2}{\sum (x_i - \bar{x})^2} \right) \space$ (for an actual value)
    
The latter has two sources of variability: variance from regression line, variance from $\epsilon$.

Prediction Interval: $\space \hat{y}^* \pm t_{n-2, \alpha/2} \sqrt{S^2 \left( 1 + \frac{1}{N} + \frac{(x^* - \bar{x})^2}{\sum (x_i - \bar{x})^2} \right)}$

## Gauss-Markov Conditions

$\mathbf{\epsilon} \sim \mathcal{N} \left( \mathbf{0}, \hspace{2mm} diag(\sigma^2) \right)$

Errors have zero mean, are homoskedastic and uncorrelated.

If satisfied, then OLS estimators are the Best Linear Unbiased Estimators ('best' implies minimum variance in estimator).

## Explained Variance

$SST = SSReg + RSS$

$\sum (y_i - \bar{y})^2 = \sum (\hat{y}_i - \bar{y})^2 + \sum (y_i - \hat{y}_i)^2$

$R^2 = \frac{SSReg}{SST} = \begin{cases} 1, \hat{y}_i = y_i \\ 0, \hat{y}_i = \bar{y} \end{cases} \forall \space i \space$ (Coefficient of Determination)

$R^2 = r^2 \space$ (Coefficient of Correlation Squared)

## Diagnostics

Standardized Residuals
* $r_i = \frac{\hat{\epsilon}_i}{\sqrt{S^2(1-h_{ii})}}$

&nbsp;

Cook's Distance: measures influence of i-th observation
* $D_i = \frac{\sum^N_{j=1} (\hat{y}_{j(i)} - \hat{y}_j)^2}{2S^2}, \space \hat{y}_{j(i)} = $ fitted value for point j, fitted on all points excluding the i-th point

* $D_i = \frac{r^2_i}{2} \left( \frac{h_{ii}}{1-h_{ii}} \right)$

i.e. both standardized residual and leverage contributes to influence

&nbsp;

* Outliers: $|r_i| > 2$ (small datasets), $|r_i| > 4$ (large datasets)

* Leverage points: outliers w.r.t. x-axis ($h_{ii} > \frac{4}{N}$)

* Influential points: removing the point changes fit drastically $(D_i > \frac{4}{N-2})$

### Quantile-Quantile (QQ) Plot

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

* Sample quantiles (y-axis) vs Theoretical quantiles (x-axis)

```python
def qq_plot(data, quantile):
    data_percentiles = np.percentile(data, q=np.arange(quantile*100,100,quantile*100))
    normal_percentiles = ss.norm.ppf(np.arange(quantile,1,quantile))
    plt.plot(normal_percentiles, data_percentiles, marker=".", ls="")
    # reference line: S.D. change in sample scale for every
    # S.D change in theoretical scale (using Robust estimation)
    iqr, median = ss.iqr(data), np.median(data)
    iqr_supp = ss.norm.ppf(0.75) - ss.norm.ppf(0.25)
    reference_line = [median + iqr/iqr_supp*x for x in normal_percentiles]
    plt.plot(normal_percentiles, reference_line)
    plt.ylim(min(data), max(data))
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Sample Quantiles")
    return plt
```

```python
uniform_data = sorted(np.random.uniform(low=0, high=1.0, size=500))
plot1 = qq_plot(uniform_data, quantile=0.01)
```
